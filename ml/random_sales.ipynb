```python
import pandas as pd
import numpy as np

dates = pd.date_range("2023-01-01", "2024-12-01", freq="D")
product_ids = [201,202,203]
rows = []

for d in dates:
    for p in product_ids:
        qty = np.random.poisson(3)
        rows.append([None, d, 1000+p, p, qty, np.random.randint(80,150)])

df = pd.DataFrame(rows, columns=["order_id","order_date","customer_id","product_id","qty","unit_price"])
df.to_csv("C:/Users/Krish/cloud-erp-lite/ml/sample_sales.csv", index=False)
```
